```python
import pandas as pd
import inflection

from FiveETools.core.fantasy.sources import full_source, source, json_source, sources
import FiveETools.core.fantasy.magic_items
import FiveETools.core.fantasy.spells
import FiveETools.core.fantasy.monster
import FiveETools.core.fantasy.conditions
import FiveETools.core.fantasy.languages
import FiveETools.core.fantasy.species
import FiveETools.core.fantasy.dieties
import FiveETools.core.fantasy.diseases

import json
import hashlib
import time


def generate_date_last_modified_hash(doc) -> str:
    if isinstance(doc, dict):
        doc = dict(doc)
        doc.pop("_dateLastModifiedHash", None)

    json_str = json.dumps(
        doc,
        ensure_ascii=False,
        sort_keys=True,
        default=lambda o: list(o) if isinstance(o, set) else o
    )

    md5_hash = hashlib.md5(json_str.encode("utf-8")).hexdigest()
    return md5_hash


current_epoch = time.time()
compendium = {
    "_meta": {
        "sources": sources,
        "unlisted": True,
        "dateAdded": 1743879729,
        "dateLastModified": int(current_epoch),
        "_dateLastModifiedHash": "??????",
        "edition": "classic",
        "currencyConversions": {
            "credit": [{"coin": "credit", "mult": 0.001, "isFallback": True}]
        },
    },
    "$schema": "https://raw.githubusercontent.com/TheGiddyLimit/5etools-utils/master/schema/brew-fast/homebrew.json",
    "race": FiveETools.core.fantasy.species.species_list,
    # "class": FiveETools.core.modern.classes.classes_list,
    # "subclass": FiveETools.core.modern.subclasses.subclasses_list,
    "spell": FiveETools.core.fantasy.spells.spells_list,
    # "item": FiveETools.core.modern.items.items_list + FiveETools.core.modern.magic_items.magic_items_list,
    "monster": FiveETools.core.fantasy.monster.monster_list,
    # "feat": FiveETools.core.modern.feats.feat_list,
    # "background": FiveETools.core.modern.backgrounds.background_list,
    "condition": FiveETools.core.fantasy.conditions.condition_list,
    "disease": FiveETools.core.fantasy.diseases.disease_list,
    "language": FiveETools.core.fantasy.languages.language_list,
    "deity": FiveETools.core.fantasy.dieties.diety_list,
    # "classFeature": FiveETools.core.modern.features.features_list,
    # "subclassFeature": FiveETools.core.modern.features.sub_class_features_list,
    # "itemProperty": FiveETools.core.modern.items.item_property_list,
}

new_hash = generate_date_last_modified_hash(compendium)
compendium["_meta"]["_dateLastModifiedHash"] = f"{new_hash[:10]}"
# Write the compendium to a JSON file
output_file = f"Velum_Cineris;{inflection.underscore("GuideToOrimond")}.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(
        compendium,
        f,
        ensure_ascii=False,
        indent=4,
        default=lambda o: list(o) if isinstance(o, set) else o
    )

print(f"Export successful! The fantasy compendium has been saved as '{output_file}'.")
```
