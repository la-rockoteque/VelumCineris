```python
import pandas as pd
import inflection

from src.sources import full_source, source, json_source, sources
import FiveETools.modern.items
import FiveETools.modern.magic_items
import FiveETools.modern.feats
import FiveETools.modern.spells
import FiveETools.modern.classes
import FiveETools.modern.subclasses
import FiveETools.modern.monster
import FiveETools.modern.conditions
import FiveETools.modern.features
import FiveETools.modern.languages
import FiveETools.modern.species
import FiveETools.modern.backgrounds
import FiveETools.modern.dieties
import FiveETools.modern.diseases

import json
import hashlib
import time


def generate_date_last_modified_hash(doc) -> str:
    if isinstance(doc, dict):
        doc = dict(doc)
        doc.pop("_dateLastModifiedHash", None)

    json_str = json.dumps(
        doc,
        ensure_ascii=False,
        sort_keys=True,
        default=lambda o: list(o) if isinstance(o, set) else o
    )

    md5_hash = hashlib.md5(json_str.encode("utf-8")).hexdigest()
    return md5_hash


current_epoch = time.time()
compendium = {
    "_meta": {
        "sources": sources,
        "unlisted": True,
        "dateAdded": 1743879729,
        "dateLastModified": int(current_epoch),
        "_dateLastModifiedHash": "??????",
        "edition": "classic",
        "currencyConversions": {
            "credit": [{"coin": "credit", "mult": 0.001, "isFallback": True}]
        },
    },
    "$schema": "https://raw.githubusercontent.com/TheGiddyLimit/5etools-utils/master/schema/brew-fast/homebrew.json",
    "race": FiveETools.modern.species.species_list,
    "class": FiveETools.modern.classes.classes_list,
    "subclass": FiveETools.modern.subclasses.subclasses_list,
    "spell": FiveETools.modern.spells.spells_list,
    "item": FiveETools.modern.items.items_list + FiveETools.modern.magic_items.magic_items_list,
    "monster": FiveETools.modern.monster.monster_list,
    "feat": FiveETools.modern.feats.feat_list,
    "background": FiveETools.modern.backgrounds.background_list,
    "condition": FiveETools.modern.conditions.condition_list,
    "disease": FiveETools.modern.diseases.disease_list,
    "language": FiveETools.modern.languages.language_list,
    "deity": FiveETools.modern.dieties.diety_list,
    "classFeature": FiveETools.modern.features.features_list,
    "subclassFeature": FiveETools.modern.features.sub_class_features_list,
    "itemProperty": FiveETools.modern.items.item_property_list,
}

new_hash = generate_date_last_modified_hash(compendium)
compendium["_meta"]["_dateLastModifiedHash"] = f"{new_hash[:10]}"
# Write the compendium to a JSON file
output_file = f"Velum_Cineris;{inflection.underscore(json_source)}.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(
        compendium,
        f,
        ensure_ascii=False,
        indent=4,
        default=lambda o: list(o) if isinstance(o, set) else o
    )

print(f"Export successful! The compendium has been saved as '{output_file}'.")
```
