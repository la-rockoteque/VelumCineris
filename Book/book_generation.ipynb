# Book Generation

Generate D&D sourcebooks in Google Docs with PHB-style formatting.

This notebook consolidates homebrew content from Google Sheets into professional-looking books.

## 1. Setup

Import required modules.

```python
# Add parent directory to path
import sys
sys.path.insert(0, '..')

from Book.core.Helpers.book_api import BookAPI
from Book.core.Helpers.google_docs_client import GoogleDocsClient
from Book.core.writers import (
    OmnibookWriter,
    PHBWriter,
    DMGWriter,
    MonsterManualWriter,
    DivineCodexWriter
)
from FiveETools.core.Helpers.gsheets_client import fantasy_sheets, modern_sheets

print("✓ Imports successful")
```

## 2. Configuration

Set the document ID and content source.

```python
# Google Docs document ID
OMNIBOOK_ID = "1_a0yx9UnrsE4oPdkDS1WSkF1nvf-L4PV4bSDKluSI-w"

# Credentials path
CREDENTIALS_PATH = "../FiveETools/key.json"

# Content source: "fantasy" (Orimond) or "modern" (Concord City)
SOURCE = "fantasy"

print(f"✓ Configuration set:")
print(f"  Document ID: {DOC_ID}")
print(f"  Source: {SOURCE}")
```

## 3. Initialize

Create the Google Docs client and BookAPI.

```python
# Initialize Google Docs client
gdocs = GoogleDocsClient(DOC_ID, CREDENTIALS_PATH)

# Select appropriate sheets client
gsheets = fantasy_sheets if SOURCE == "fantasy" else modern_sheets

# Initialize BookAPI
book_api = BookAPI(gdocs, gsheets)

print("✓ API clients initialized")
```

## 4. Preview Section (Optional)

Preview formatting for a specific entity type before generating the full book.

```python
# Preview spells (change entity_type as needed)
book_api.preview_section("spell", source=SOURCE, limit=3)
```

## 5. Generate Omnibook (Complete)

Generate a complete book with all entity types.

```python
# Create Omnibook writer
writer = OmnibookWriter(book_api, source=SOURCE)

# Generate book
book_api.generate_book(writer, DOC_ID)

print("\n✓ Book generation complete!")
print(f"View at: https://docs.google.com/document/d/{DOC_ID}/edit")
```

## 6. Generate Player's Handbook

Generate a PHB-style book with player-facing content only.

```python
# Create PHB writer
writer = PHBWriter(book_api, source=SOURCE)

# Generate book
book_api.generate_book(writer, DOC_ID)

print("\n✓ PHB generation complete!")
print(f"View at: https://docs.google.com/document/d/{DOC_ID}/edit")
```

## 7. Generate Monster Manual

Generate a Monster Manual with creatures only.

```python
# Create Monster Manual writer
writer = MonsterManualWriter(book_api, source=SOURCE)

# Generate book
book_api.generate_book(writer, DOC_ID)

print("\n✓ Monster Manual generation complete!")
print(f"View at: https://docs.google.com/document/d/{DOC_ID}/edit")
```

## 8. Generate Dungeon Master's Guide

Generate a DMG-style book with DM-facing content.

```python
# Create DMG writer
writer = DMGWriter(book_api, source=SOURCE)

# Generate book
book_api.generate_book(writer, DOC_ID)

print("\n✓ DMG generation complete!")
print(f"View at: https://docs.google.com/document/d/{DOC_ID}/edit")
```

## 9. Load Specific Entity Types

Test loading specific entity types from Google Sheets.

```python
# Load and display entity counts
entity_types = ["spell", "monster", "species"]

for entity_type in entity_types:
    try:
        entities = book_api.load_entities(entity_type, source=SOURCE)
        print(f"✓ {entity_type.capitalize()}: {len(entities)} loaded")
    except Exception as e:
        print(f"✗ {entity_type.capitalize()}: Error - {e}")
```

## Usage Tips

- **Document ID**: Change `DOC_ID` to target different Google Docs
- **Source**: Set `SOURCE` to "fantasy" or "modern" for different campaign settings
- **Preview**: Use `book_api.preview_section()` to test formatting before full generation
- **Credentials**: Ensure `key.json` exists at the specified path

## Troubleshooting

- **Import errors**: Ensure you're running from the `Book/` directory
- **Auth errors**: Check that `key.json` has Google Docs API permissions
- **Missing entities**: Verify entity converters exist in `FiveETools/`
- **Formatting issues**: Check formatter implementations in `Book/core/formatters/`
