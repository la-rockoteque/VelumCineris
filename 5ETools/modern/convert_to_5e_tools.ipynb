```python
import pandas as pd
import inflection

from src.sources import full_source, source, json_source, sources
import 5ETools.modern.items
import 5ETools.modern.magic_items
import 5ETools.modern.feats
import 5ETools.modern.spells
import 5ETools.modern.classes
import 5ETools.modern.subclasses
import 5ETools.modern.monster
import 5ETools.modern.conditions
import 5ETools.modern.features
import 5ETools.modern.languages
import 5ETools.modern.species
import 5ETools.modern.backgrounds
import 5ETools.modern.dieties
import 5ETools.modern.diseases

import json
import hashlib
import time


def generate_date_last_modified_hash(doc) -> str:
    if isinstance(doc, dict):
        doc = dict(doc)
        doc.pop("_dateLastModifiedHash", None)

    json_str = json.dumps(
        doc,
        ensure_ascii=False,
        sort_keys=True,
        default=lambda o: list(o) if isinstance(o, set) else o
    )

    md5_hash = hashlib.md5(json_str.encode("utf-8")).hexdigest()
    return md5_hash


current_epoch = time.time()
compendium = {
    "_meta": {
        "sources": sources,
        "unlisted": True,
        "dateAdded": 1743879729,
        "dateLastModified": int(current_epoch),
        "_dateLastModifiedHash": "??????",
        "edition": "classic",
        "currencyConversions": {
            "credit": [{"coin": "credit", "mult": 0.001, "isFallback": True}]
        },
    },
    "$schema": "https://raw.githubusercontent.com/TheGiddyLimit/5etools-utils/master/schema/brew-fast/homebrew.json",
    "race": 5ETools.modern.species.species_list,
    "class": 5ETools.modern.classes.classes_list,
    "subclass": 5ETools.modern.subclasses.subclasses_list,
    "spell": 5ETools.modern.spells.spells_list,
    "item": 5ETools.modern.items.items_list + 5ETools.modern.magic_items.magic_items_list,
    "monster": 5ETools.modern.monster.monster_list,
    "feat": 5ETools.modern.feats.feat_list,
    "background": 5ETools.modern.backgrounds.background_list,
    "condition": 5ETools.modern.conditions.condition_list,
    "disease": 5ETools.modern.diseases.disease_list,
    "language": 5ETools.modern.languages.language_list,
    "deity": 5ETools.modern.dieties.diety_list,
    "classFeature": 5ETools.modern.features.features_list,
    "subclassFeature": 5ETools.modern.features.sub_class_features_list,
    "itemProperty": 5ETools.modern.items.item_property_list,
}

new_hash = generate_date_last_modified_hash(compendium)
compendium["_meta"]["_dateLastModifiedHash"] = f"{new_hash[:10]}"
# Write the compendium to a JSON file
output_file = f"Velum_Cineris;{inflection.underscore(json_source)}.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(
        compendium,
        f,
        ensure_ascii=False,
        indent=4,
        default=lambda o: list(o) if isinstance(o, set) else o
    )

print(f"Export successful! The compendium has been saved as '{output_file}'.")
```
